```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def draw_line_plot():
# Read the data and set the index to the date column
 df = pd.read_csv('fcc-forum-pageviews.csv', parse_dates=['date'], index_col='date')
# Clean the data by filtering out days when the page views were in the top 2.5% or bottom 2.5% of the dataset
df_cleaned = df[(df['value'] >= df['value'].quantile(0.025)) & (df['value'] <= df['value'].quantile(0.975))]
# Draw line plot
plt.figure(figsize=(10, 5))
plt.plot(df_cleaned.index, df_cleaned['value'], color='violet')
plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
plt.xlabel('Date')
plt.ylabel('Page Views')
plt.grid(True)
plt.show()
```

```python
def draw_bar_plot():
# Read the data and set the index to the date column
    df = pd.read_csv('fcc-forum-pageviews.csv', parse_dates=['date'], index_col='date')
 # Clean the data by filtering out days when the page views were in the top 2.5% or bottom 2.5% of the dataset
df_cleaned = df[(df['value'] >= df['value'].quantile(0.025)) & (df['value'] <= df['value'].quantile(0.975))]
# Avoid SettingWithCopyWarning by explicitly copying the DataFrame
df_cleaned = df_cleaned.copy()
# Ensure correct grouping by 'year' and 'month'
df_cleaned.loc[:,'year'] = df_cleaned.index.year
df_cleaned.loc[:,'month'] = df_cleaned.index.month
# Group by year and month and calculate the average daily page views
df_grouped = df_cleaned.groupby(['year', 'month']).mean().reset_index()
# Pivot the DataFrame to have years as columns and months as rows
df_pivot = df_grouped.pivot(index='year',columns='month',values='value')
# Draw bar plot
plt.figure(figsize=(10, 5))
sns.barplot(data=df_grouped,x='year',y='value',hue='month',errorbar=None)
plt.title('Average Daily Page Views')
plt.xlabel('Years')
plt.ylabel('Average Page Views')
plt.legend(title='Months', labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
def draw_box_plot():
# Read the data and set the index to the date column
 df = pd.read_csv('fcc-forum-pageviews.csv', parse_dates=['date'], index_col='date')
# Clean the data by filtering out days when the page views were in the top 2.5% or bottom 2.5% of the dataset
df_cleaned = df[(df['value'] >= df['value'].quantile(0.025)) & (df['value'] <= df['value'].quantile(0.975))]
# Avoid SettingWithCopyWarning by explicitly copying the DataFrame
df_cleaned = df_cleaned.copy()
# Prepare data for box plots
df_cleaned.loc[:,'year'] = df_cleaned.index.year
df_cleaned.loc[:,'month'] = df_cleaned.index.strftime('%b')
df_cleaned.loc[:,'month_num'] = df_cleaned.index.month
# Draw box plots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
sns.boxplot(data=df_cleaned, x='year', y='value', ax=axes[0])
axes[0].set_title('Year-wise Box Plot (Trend)')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Page Views')
sns.boxplot(data=df_cleaned, x='month_num', y='value', ax=axes[1])
axes[1].set_title('Month-wise Box Plot (Seasonality)')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Page Views')
axes[1].set_xticks(range(1,13))
axes[1].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
